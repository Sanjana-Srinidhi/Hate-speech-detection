```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {'XGBoost': 0.74, 'Logistic Regression': 0.79, 'KNN': 0.65, 'Decision Trees': 0.63, 'Random Forests': 0.79, 'Linear SVM': 0.77, 'Adaboost': 0.75, 'Passive Aggressive Classifier': 0.76, 'Naive Bayes Classifier': 0.81}
models = list(data.keys())
acc = list(data.values())
```

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(models, acc, color = 'blue', width = 0.4)
plt.xticks(rotation = 90)
plt.xlabel("ML Models")
plt.ylabel("Accuracy")
plt.show()
```
