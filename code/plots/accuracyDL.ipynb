```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {'FastText+BiSLTM': 0.65, 'GloVe+BiLSTM': 0.64, 'RoBerta': 0.51}
models = list(data.keys())
acc = list(data.values())
```

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(models, acc, color = 'blue', width = 0.4)
plt.xlabel("DL Models")
plt.ylabel("Accuracy")
plt.show()
```
